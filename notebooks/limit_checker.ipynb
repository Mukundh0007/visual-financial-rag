```python
def print_available_creds():
    api_key = os.getenv("OPENROUTER_API_KEY")
    response = requests.get(
        url="https://openrouter.ai/api/v1/key",
        headers={
            "Authorization": f"Bearer {api_key}"
    })
    available_creds = float(json.dumps(response.json()["data"]["limit_remaining"]))
    print("AVAILABLE CREDITS =",available_creds)
```
